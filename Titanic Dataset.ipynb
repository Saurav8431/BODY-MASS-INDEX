```python
df = pd.read_csv("Titanic-Dataset.csv")

# Joint probability distribution of 3 variables
joint_prob = pd.crosstab(
    [df['Pclass'], df['Sex']],
    df['Survived'],
    normalize=True
)

print(joint_prob)
```
